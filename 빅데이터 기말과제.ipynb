```python
death_df=pd.read_csv('C://Users//RB-N8//Desktop//12_20_death.csv', encoding='cp949')
all_data = { # 필요한 정보를 추가함
    'year': death_df['발생년'],
    'M_Day': death_df['발생년월일시'],
    'DaN': death_df['주야'],
    'day': death_df['요일'],
    'death': death_df['사상자수'],
    'sido': death_df['발생지시도'],
    'gu': death_df['발생지시군구'],
    'Law': death_df['법규위반'],
    'Car': death_df['당사자종별_1당_대분류'],
    'Lat': death_df['경도'],
    'Lng': death_df['위도'],
    'DH': death_df['사망자수'],
    'MH': death_df['중상자수'],
    'WH': death_df['경상자수']
}

data = pd.DataFrame(all_data) # DataFrame 형식으로 변환
data # 확인을 위한 출력
```

```python
from sklearn.linear_model import LinearRegression # 선형회귀를 위해 호출
from sklearn.model_selection import train_test_split # 학습을 위해 호출
from sklearn.metrics import mean_squared_error, r2_score # 평균 제곱 오차를 위한 호출
```

```python
# 사망자, 중상자, 경상자, 위도, 경도를 제외한 나머지 컬럼을 독립변수에서 제외
X = data.drop(['year', 'M_Day', 'DaN', 'day', 'sido', 'gu', 'Law', 'Car'], axis=1, inplace=False)
Y = data['death'] # 사상자 수를 결과값으로 설정
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state=2) # X, Y의 학습, 테스트 데이터
```

```python
lr = LinearRegression() # 선형 회귀 변수
lr.fit(X_train, Y_train) # 학습시킨다
print(lr.score(X_test, Y_test)) # 출력
```

```python
Y_predict = lr.predict(X_test) # Y의 예상 데이터
print(Y_predict) # 출력
```

```python
mse = mean_squared_error(Y_test, Y_predict) # Y의 학습, 예측 평균 제곱 오차를 구한다
rmse = np.sqrt(mse) # 평균 제곱 오차의 제곱근
print(mse) # 평균 제곱 오차 출력
print(rmse) # 평균 제곱 오차의 제곱급 출력
print(r2_score(Y_test, Y_predict))
```

```python
print(lr.coef_) # 선형 회귀 기울기
print(lr.intercept_) # 선형 회귀 절편
print(np.round(lr.coef_,1))
```

```python
coef = pd.Series(data = np.round(lr.coef_, 1), index = X.columns) # 회귀 계수를 큰 값 순으로 정렬하기 위해 Series로 생성
coef.sort_values(ascending = False) # 정렬
```

```python
# 연도에 따라 다른 색을 표현하기 위해 메소드 제작
def color_select(year):
    if year== 2012:
        return 'white'
    elif year== 2013:
        return 'orange'
    elif year== 2014:
        return 'yellow'
    elif year== 2015:
        return 'green'
    elif year== 2016:
        return 'blue'
    elif year== 2017:
        return 'Indigo'
    elif year== 2018:
        return 'purple'
    elif year== 2019:
        return '#BF2F7B'
    else:
        return '#BD7F41'
```

```python
import folium

# 처음 지도의 시작 위치를 입력하고 생성
accident_circle = folium.Map(location = [37.53897093698831, 127.4935922], zoom_start=14)

# 지도에 원형 마커를 여러 개 찍음
for lat, lng, year in zip(chungbuk['Lat'], chungbuk['Lng'], chungbuk['year']):
    folium.CircleMarker(
        [lng, lat],
        radius=3,
        color= color_select(year)
    ).add_to(accident_circle)
    
accident_circle
```
